```python


# URL dos dados
url = "https://servicodados.ibge.gov.br/api/v3/agregados/8687/periodos/2015|2016|2017|2018|2019|2020|2021/variaveis/6843|6846?localidades=N1[all]"

# Fazer uma solicitação GET para obter os dados
resposta = requests.get(url)

# Verificar o status da resposta
if resposta.status_code == 200:
    data = resposta.json()
else:
    print("Falha ao obter os dados. Verifique a URL e a conexão com a internet.")
    exit()

# Converter os dados em um DataFrame do Pandas
df = pd.DataFrame(data)

```

```python
# Realizar análise exploratória de dados
# Aqui você pode realizar várias análises, como estatísticas descritivas e visualizações
# Por exemplo, você pode usar o matplotlib para criar gráficos

# Histograma
plt.figure(figsize=(10, 6))
plt.hist(df['variavel'], bins=20, color='blue')
# plt.hist(df['unidade'], bins=20, color='green')
plt.title('Histograma da Disponibilidade de Rede de Água')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()

```

```python
# Box plot
plt.figure(figsize=(10, 6))
plt.boxplot(df['variavel'])
plt.title('Box Plot da Disponibilidade de Rede de Água')
plt.ylabel('Valor')
plt.show()

    
```
